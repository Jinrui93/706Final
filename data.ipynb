```python
import pandas as pd
```

```python
df = pd.read_csv('covid.csv')
```

```python
df.head()
```

```python
df1 = df[['date','county','state','cases','deaths','percent_uninsured','percent_vaccinated']]
df1.head()
```

```python
df2 = df1.groupby(['date','state']).agg({'cases':['sum'], 'deaths':['sum'], 'percent_uninsured':['median'], 'percent_vaccinated':['median']})
df2.tail(20)
```

```python
df3 = df2.dropna().reset_index()
df3.columns = df3.columns.droplevel(1)
df3.tail(20)
```

```python
df3['date'] = pd.to_datetime(df3['date'])
df3['new_date'] = df3['date'].dt.strftime('%Y%m')
df3.tail(20)
```

```python
df4 = df3.groupby(['new_date','state']).agg({'cases':['sum'], 'deaths':['sum'], 'percent_uninsured':['median'], 'percent_vaccinated':['median']})
df4.tail(20)
```

```python
df4['cases_k'] = round(df4['cases']/1000,2)
df4['deaths_k'] = round(df4['deaths']/1000,2)
df4['percent_uninsured'] = round(df4['percent_uninsured'],2)
df4.tail(20)
```

```python
df5 = df4.dropna().reset_index()
df5.columns = df5.columns.droplevel(1)
df5.tail(20)
```

```python
df5['percent_insured'] = 100-df5['percent_uninsured']
df5.tail(20)
```

```python
df5.to_csv('data_month.csv', index=False)
```
